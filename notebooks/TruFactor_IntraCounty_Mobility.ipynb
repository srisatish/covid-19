# TruFactor COVID-19 Geo Mobility
<p>TruFactor’s Geographic Mobility Dataset aims to provide a national and regional understanding of mobility changes across a significant portion of the United States while maintaining individual privacy. This dataset is intended to enable researchers, public officials and citizens to understand how population level behaviors are changing due to COVID-19 and the ensuing national, state and local response. 

Two datasets are being produced which may be used independently or in tandem to analyze how the population level mobility behaviors have changed. The first dataset provides intercounty travel patterns at a daily frequency. <b>TruFactor's InterCounty Mobility</b> dataset produces estimates of where a resident population is traveling to. This dataset can aid in understanding both the export and import flows of potentially infected populations. An example use of this dataset would be to model how a virus could spread overtime. Another potential use could be for modeling economic impacts due to reduced mobility patterns.

The second dataset provides intracounty estimates of how mobility behaviors are changing by demographic group. <b>TruFactor's IntraCounty Mobility</b> dataset measures the change in travel behavior for a county’s residents by providing a decile distribution of the # of visited census tracts within the county. Travel behaviors for residents within a county by demographic group are provided to show how mobility is changing by age group at a daily frequency. Researchers and policymakers can understand how behaviors are changing in response to infection rates and policy shocks such as Stay At Home policies.
    
In this tutorial we will show you how to work with free samples of <b>IntraCounty Mobility</b> that a published in Amazon Data Exchange (https://aws.amazon.com/data-exchange/). We walk through several applications and show how some intelligent applications of these intelligence sets can power vertical oriented applications.    
</p>

## Step 1: Measuring Social Distancing
<p> An ideal measure of social distancing would measure a populations change in mobility behavior over time. To this end, TruFactor's InterCounty Mobility aims to measure the change in social distancing through distributional changes in intracounty travel patterns. Specifically, we measure the number of census tracts visited by the resident county population. We posit changes in the distribution are indicative of changes in the populations mobility and thus a macro measure of social distancing.
</p>

<p>
Furthermore, we suggest a simple measure for measuring the change in the distribution. Area under the curve (AUC) measures the total area between an axis and some line segment. In this instance, AUC changes can be a proxy for how large the change in social distancing is.</p>

```python
intra_path = 's3://trufactor-demos/covid/IntraCountyMobility'
intra = spark.read.parquet(intra_path)
intra.printSchema()
```

```python
intra.show(1)
```

```python
intra_ex = (intra.select('HomeCounty',
              'AgeGroup',
              fns.posexplode('VisitDeciles').alias('decile','visits'),
              'dt'
             )
            .filter("dt in (20200219,20200304, 20200318)")
         )

test = intra_ex.filter("HomeCounty='24027'").toPandas()
```

```python
sns.set_context('notebook')
g = sns.FacetGrid(test,col='AgeGroup', hue='dt', col_wrap=2,  height=4, aspect=1.5)
g.map(sns.lineplot,'decile','visits').add_legend()
g.fig.suptitle("Census Tract Visitation by Age Category for County FIPS 24027",y=1)
g.fig.tight_layout(pad=2,h_pad=10)
```

<br>
<p> In the above plot, it is clear that 2/19 represented a peak of sorts and that subsequent Wednesdays have seen declines. By March 18th, most of the deciles across all age groups experience a distinct shift down and to the right. Interestingly, we see a flattening of the curve across the first several deciles. If the 2nd decile is 1, then 30% of people did not visit more than a single census tract that day. Basically, they stayed home!
</p>

<p> To compute a single measure which we can compare, a researcher can simply calculate the AUC of each curve. This can be done simply by summing the visits across the deciles. Then comparing the change from across two time intervals. We'll compare the 9 largest counties in the US to see how they are doing at social distancing.
    
    County data can be downloaded here: https://www2.census.gov/programs-surveys/popest/datasets/2010-2019/counties/totals/co-est2019-alldata.csv
</p>

```python
co = pd.read_csv('co-est2019-alldata.csv',encoding='latin-1')
```

```python
co.STATE = co.STATE.map(lambda x: str(x).zfill(2))
co.COUNTY = co.COUNTY.map(lambda x: str(x).zfill(3))
co['pop'] = co.POPESTIMATE2019
co['county'] = co.STATE+co.COUNTY

copop = spark.createDataFrame(co[co.STNAME!=co.CTYNAME][['county','pop']])
```

```python
w = Window().partitionBy('HomeCounty','AgeGroup').orderBy('dt').rowsBetween(Window.unboundedPreceding, Window.unboundedFollowing)
```

```python
intra_grp = (intra_ex
    .groupby('AgeGroup','dt','HomeCounty')
    .agg(fns.sum('visits').alias('visits'))
)

z = (intra_grp.orderBy('HomeCounty','AgeGroup','dt').select('*',fns.last('visits').over(w).alias('last_visit'))
    .filter('dt=20200219')
    .select('*',
            ((col('last_visit')-col('visits'))/col('visits')).alias('pct_change')
            )
    .orderBy(col('pct_change').asc())
     .withColumnRenamed('HomeCounty','county')

)
```

```python
big = copop.orderBy(col('pop').desc()).limit(10)
d = z.join(big,'county').toPandas()
```

```python
clkp = pd.read_csv('county_fips.csv',dtype=str)
clkp['FIPS'] = clkp.FIPS.map(lambda x: x.zfill(5))
clkp['HomeCounty'] = clkp.County+', '+clkp.State

d = d.merge(clkp,left_on='county',right_on='FIPS')
```

```python
sns.set_context('talk',font_scale=.7)
g = sns.FacetGrid(d, col='HomeCounty',palette="Set3",col_wrap=3, height=4, aspect=1.5, sharex=False)
g.map(sns.barplot,"AgeGroup",'pct_change',**{'order':d.AgeGroup.unique()[:-1]})
g.fig.suptitle("Change in Mobility: 2-19-20 compared to 3-18-20",y=1)
g.fig.tight_layout(pad=4)
g.set_xticklabels(rotation=20)
```
